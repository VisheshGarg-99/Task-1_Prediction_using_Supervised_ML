```python
# importing the required libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Reading the Data
data = pd.read_csv ('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
data.head()
```

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
```

```python
# Check if there any null value in the Dataset
data.isnull().sum()
```

```python
plt.scatter(X,Y)
plt.xlabel("Study Hours")
plt.ylabel("Study Marks")
plt.title("Study Hours Vs Student Marks ")
plt.show()
```

**From the above scatter plot there looks to be correlation between the 'Marks Percentage' and 'Hours Studied', Lets plot a regression line to confirm the correlation.

# Training the Model

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3, random_state=0)
```

```python
print("shape of X Train",X_train.shape)
print("shape of Y Train",Y_train.shape)
print("shape of X Test", X_test.shape)
print("shape of Y Test",Y_test.shape)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_test,Y_test)
```

```python
plt.scatter(X_test,Y_test)
```

```python
plt.xlabel("Study Hours")
plt.ylabel("Study Marks")
plt.scatter(X_test,Y_test)
plt.plot(X_train,lr.predict(X_train),color="red")
```

# predicted score of a student if he/she studies for 9.25 hrs/ day?

```python
pre = lr.predict([[9.25]])[0].round(1)
print('the predicted value if the study hour is 9.25 :', pre)
```

# According to the regression model if a student studies for 9.25 hours a day he/she is likely to score 92.9 marks.
